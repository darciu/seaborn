```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)

#tips dataset (it can be either loaded as two further datasets)
url="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"
df_tips=pd.read_csv(url)
#iris dataset
df_iris = sns.load_dataset("iris")
#fmri dataset
df_fmri = sns.load_dataset("fmri")
#titanic dataset
df_titanic = sns.load_dataset('titanic')
```

## Statistical Relationships
Visualization of statistical relationships allows to see how variables in datasets depend on each other.

Iris' dataset species has different colors and markers. It may be treated like a third dimension on relplot. <br>
Below, Tips dataset is divided by categorical column 'time' and presented on two subplots. Moreover third dimension is presented by integer column 'size'.

```python
sns.relplot(x="sepal_length", y="petal_length", hue="species", style ="species", data=df_iris)
```

```python
sns.relplot(x="total_bill", y="tip", size="size", sizes=(15, 200), col = "time",color = "y", data=df_tips)
```

Example of linear relplot<br>
(_it can be obtained either with a lineplot_)

```python
sns.relplot(x="timepoint",y="signal", kind = "line", data = df_fmri)
```

## Categorical plots<br>
This type of plot generally can be obtained with 'catplot' type of seaborn plot. One of axes has to be numeric. Below are some examples

```python
sns.catplot(x ='smoker', y = 'tip', data = df_tips)
```

Swarm type

```python
sns.catplot(x='class', y='fare',hue = 'survived', kind='swarm', data = df_titanic)
```

## Distribution

Boxplot type (distribution of data). Useful when there is a need to see if data has outliers.

```python
sns.catplot(x = 'fare', kind ='box', data = df_titanic)
```

Let's exclude fare values above 100 and see boxplots for sexes and classes

```python
sns.catplot(x = 'fare',y ='sex', row = 'class', kind ='box', data = df_titanic[df_titanic.fare < 100])
```

Countplot example

```python
sns.countplot(x = 'class', order = ['Third', 'Second','First'], data = df_titanic)
```

Joint Plot graph with two variables, regression, their correlation and distribution

```python
sns.jointplot(x = 'total_bill', y = 'tip', kind = 'reg', color = 'b', data = df_tips)
```

Distplots for normal and uniform distributions

```python
x = np.random.normal(size = 1500)
sns.distplot(x, bins = 50)
```

```python
x = np.random.randint(1, 100, size = 1500)
sns.distplot(x, bins = 25)
```
